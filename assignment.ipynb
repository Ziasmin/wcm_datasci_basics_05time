```python
# importing libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
#slide 2
pi = np.pi 
f = 2 
a = 2 
T = 2 
fs = 200 

time_vector = np.linspace(0, T, fs*T)
y = a*np.sin(2*pi*f*time_vector)
rows = 2 
cols = 2

axs = plt.figure(figsize = (20,10)).subplots(rows, cols)
plt.subplot(rows, cols, 1)
plt.plot(time_vector, y)
plt.title('2Hz Sine Wave')
plt.xlabel('time (s)')
plt.autoscale(enable=True, axis = 'x', tight=True)

fs_sampled = 5
tvec_sampled = np.linspace(0, T, T*fs_sampled) 
y_sampled=a*np.sin(2*pi*f*tvec_sampled)

plt.subplot(rows, cols, 2)
markerline, stemlines, baseline = plt.stem(tvec_sampled, y_sampled, markerfmt = 'C2o', basefmt = 'C3-')
markerline.set_markerfacecolor('none')
plt.title('Sampled')
plt.xlabel('time (s)')
plt.autoscale(enable=True, axis = 'x', tight=True)

plt.subplot(rows, cols, 3)
y_discrete = np.round(a*np.sin(2*pi*f*time_vector))
plt.plot(time_vector, y_discrete)
plt.title('Discretized')
plt.xlabel('time (s)')
plt.autoscale(enable=True, axis = 'x', tight=True) 
              
plt.subplot(rows, cols, 4)
plt.plot(time_vector, y, 'g')
markerline, stemlines, baseline = plt.stem(tvec_sampled, np.round(y_sampled))
plt.stem(tvec_sampled, np.round(y_sampled))
plt.title('Sampled and Discretized')
plt.xlabel('time (s)')
plt.autoscale(enable=True, axis = 'x', tight=True) 

plt.tight_layout()
```

```python
#slide 3
P = plt.imread('penny.jpg').copy() 

rows = 2
cols = 3

def rbg2gray(rgb):
    coeffs = np.array([0.299, 0.589, 0.114])
    return np.dot(rgb, coeffs)

axs = plt.figure(figsize = (10,10)).subplots(rows, cols)
plt.set_cmap(plt.get_cmap('copper'))
PG = rbg2gray(P) 

#2 
plt.subplot(rows, cols, 1)
plt.imshow(np.floor(PG/128))
plt.axis('off')
plt.title('2 gray levels')

#4 
plt.subplot(rows, cols, 2)
plt.imshow(np.floor(PG/64))
plt.axis('off')
plt.title('4 gray levels')

#8 
plt.subplot(rows, cols, 3)
plt.imshow(np.floor(PG/32))
plt.axis('off')
plt.title('8 grey levels')

#16 
plt.subplot(rows, cols, 4)
plt.imshow(np.floor(PG/16))
plt.axis('off')
plt.title('16 gray levels')

#32 
plt.subplot(rows, cols, 5)
plt.imshow(np.floor(PG/8))
plt.axis('off')
plt.title('32 gray levels')

#256 
plt.subplot(rows, cols, 6)
plt.imshow(np.floor(PG))
plt.axis('off')
plt.title('256 gray levels')
```

```python
#Slide 11
#Bad resolution

a = np.linspace(0, 40, 500)
b = np.linspace(0, 40/3, 200)

[x,y] = np.meshgrid(a,b) 

r2 = x**2 + y**2
z = 255*np.abs(np.sin(r2)) 

plt.figure(figsize=(10,8))
plt.imshow(z, cmap=plt.get_cmap('gray'))
plt.axis('off')
plt.title('Bad resolution')

#Good resolution
#resolution axes
sf = 2
a_sf = np.linspace(0, 9, 720*sf)
b_sf = np.linspace(0, 9/3, 240*sf)

[x_sf,y_sf] = np.meshgrid(a_sf,b_sf) 

r2_sf = x_sf**2 + y_sf**2
z_sf = 255*np.abs(np.sin(r2_sf)) #image

plt.figure(figsize=(10,8))
plt.imshow(z_sf, cmap=plt.get_cmap('gray'))
plt.axis('off')
plt.title('Good resolution')
```
